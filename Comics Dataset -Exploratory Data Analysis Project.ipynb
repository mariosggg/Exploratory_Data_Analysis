Converting Unwanted Datatypes



```python
df['APPEARANCES'] = df['APPEARANCES'].astype('Int64')

df1['APPEARANCES'] = df1['APPEARANCES'].astype('Int64')
```

Switching to Uppercase letters



```python
df.columns = df.columns.str.upper()

df1.columns = df1.columns.str.upper()
```

Getting Rid of Rows With Null Values in 'APPEARANCES' and 'FIRST APPEARANCE' Columns



```python
df = df[~df['APPEARANCES'].isnull()]
  
df1 = df1[~df1['APPEARANCES'].isnull()]

df = df[~df['FIRST APPEARANCE'].isnull()]
  
df = df1[~df1['FIRST APPEARANCE'].isnull()]
```

Arranging VERSE Column to 3rd Position



Deleting DC's Wrong Values



```python
df = df[~df['FIRST APPEARANCE'].str.contains('Holiday')]
```

Customizing Columns Values



```python
df['SEX'] = df['SEX'].str.split(' ', expand=True)[0]
df['ALIGN'] = df['ALIGN'].str.split(' ', expand=True)[0]
df['EYE'] = df['EYE'].str.split(' ', expand=True)[0]
df['HAIR'] = df['HAIR'].str.split(' ', expand=True)[0]
df['ALIVE'] = df['ALIVE'].str.split(' ', expand=True)[0]
```

```python
df1['SEX'] = df1['SEX'].str.split(' ', expand=True)[0]
df1['ALIGN'] = df1['ALIGN'].str.split(' ', expand=True)[0]
df1['EYE'] = df1['EYE'].str.split(' ', expand=True)[0]
df1['HAIR'] = df1['HAIR'].str.split(' ', expand=True)[0]
df1['ALIVE'] = df1['ALIVE'].str.split(' ', expand=True)[0]
```

Dropping Unuseful Columns



```python
#drop 'year' column in both because it's the same as 'first appearance' column
df1.drop(columns='YEAR', inplace=True)
df.drop(columns='YEAR',inplace=True)

#drop 'urlslug' column in both because the url isn't working
df1.drop(columns='URLSLUG', inplace=True)
df.drop(columns='URLSLUG',inplace=True)
```

Visualizing the Distribution of Characters by Alignment



```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

sns.countplot(x='SEX', data=df, palette='Blues_d', ax=axes[0])
axes[0].set_title('DC - Distribution of Characters by Gender')

sns.countplot(x='SEX', data=df1, palette='Reds_d' ,ax=axes[1])
axes[1].set_title('Marvel - Distribution of Characters by Gender')


plt.tight_layout()
plt.show()
```

Visualizing a Ralationship Between Alignment and Gender



```python
combined_data = pd.concat([df[['ALIGN', 'SEX']], df1[['ALIGN', 'SEX']]])

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))


sns.countplot(x='ALIGN', hue='SEX', data=df1, palette='Blues_d', ax=axes[0])
axes[0].set_title('DC - Relationship Between Alignment and Gender')
axes[0].set_xlabel('Alignment')
axes[0].set_ylabel('Count')


sns.countplot(x='ALIGN', hue='SEX', data=df, palette='Reds_d', ax=axes[1])
axes[1].set_title('Marvel - Relationship Between Alignment and Gender')
axes[1].set_xlabel('Alignment')
axes[1].set_ylabel('Count')


plt.tight_layout()
plt.show()
```

Visualizing a Relationship Between Identity and Gender

Visualizing a Relationship Between Identity and Alignment



```python
combined_data = pd.concat([df1[['ID', 'SEX']], df[['ID', 'SEX']]])

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 7))


sns.countplot(x='ID', hue='ALIGN', data=df, palette='Blues_d', ax=axes[0])
axes[0].set_title('DC - Relationship Between Alignment and Identity')
axes[0].set_xlabel('Identity')
axes[0].set_ylabel('Count')


sns.countplot(x='ID', hue='ALIGN', data=df1, palette='Reds_d', ax=axes[1])
axes[1].set_title('Marvel - Relationship Between Alignment and Identity')
axes[1].set_xlabel('Identity')
axes[1].set_ylabel('Count')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Visualizing a HeatMap for Eye and Hair Color Correlation



```python
combined_data = pd.concat([df1[['EYE', 'HAIR']], df[['EYE', 'HAIR']]])

cross_tab = pd.crosstab(combined_data['EYE'], combined_data['HAIR'])

plt.figure(figsize=(12, 8))
sns.heatmap(cross_tab, cmap='coolwarm', annot=True, fmt='d')

plt.title('Eye Color and Hair Color Correlation in Marvel and DC Characters')
plt.xlabel('Hair Color')
plt.ylabel('Eye Color')

plt.show()
```

visualize the distribution of characters based on their "Alive" status

```python
# Filter data for Marvel and DC characters based on their name or any other identifier
marvel_characters = df[df['NAME'].str.contains('Marvel', case=False)]
dc_characters = df[df['NAME'].str.contains('DC', case=False)]

# Create subplots for two pie charts
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Pie chart for Alive status in Marvel
alive_counts_marvel = marvel_characters['ALIVE'].value_counts()
axs[0].pie(alive_counts_marvel, labels=alive_counts_marvel.index, autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'])
axs[0].set_title('Distribution of Alive Characters in Marvel')

# Pie chart for Alive status in DC
alive_counts_dc = dc_characters['ALIVE'].value_counts()
axs[1].pie(alive_counts_dc, labels=alive_counts_dc.index, autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'])
axs[1].set_title('Distribution of Alive Characters in DC')

plt.tight_layout()
plt.show()

# Create subplots for two pie charts
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Pie chart for Appearance count in Marvel
axs[0].pie(marvel_characters['APPEARANCES'], autopct='%1.1f%%', colors=['skyblue'])
axs[0].set_title('Distribution of Appearances in Marvel')

# Pie chart for Appearance count in DC
axs[1].pie(dc_characters['APPEARANCES'], autopct='%1.1f%%', colors=['gold'])
axs[1].set_title('Distribution of Appearances in DC')

plt.tight_layout()
plt.show()
```
